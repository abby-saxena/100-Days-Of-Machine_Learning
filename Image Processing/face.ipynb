```python
frame = cv2.imread("im.jpg")
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces [0]
```

```python
x,y,w,h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb, (x, y), (x+w, y+h), (255,0,0), 4)
```

```python
plt.imshow(out)
```
