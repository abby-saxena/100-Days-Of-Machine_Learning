```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X = np.arange(10)
print(X)
```

```python
Y = (X-5)**2    # Y is just function of X , both x & y are numpy arrays
print(Y)
```

# Visualisation

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

# General Gradient Descent

```python
x = 0
lr = 0.1 

error = []
# 50 steps in downhill direction

plt.plot(X,Y)

for i in range(50):
    grad = 2*(x-5)               # gradient function is just diffrentiation of y function
    x = x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)              # error is same as y
    print(x)                     # at the end x is close to 5 ie our algorithm is converging
    
```

```python
# Plot the values of error - how does f(x) or error change over time

plt.plot(error)
```
