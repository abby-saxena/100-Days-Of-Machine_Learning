```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=400, n_informative=1, n_features=1, noise=1.8, random_state=11)

print(X.shape)
print(Y.shape)
```

```python
# we have to reshape Y
Y = Y.reshape((-1,1))
print(Y.shape)
```

```python
# Normalisation
X = (X-X.mean())/X.std()

# visualise

plt.figure()
plt.scatter(X,Y)
plt.title("normalised data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_.shape)
```

```python
def predict(X,theta):
    return np.dot(X,theta)


# getting theta using closed form
def get_theta(X,Y):
    
    Y = np.mat(Y)
    firstpart = np.dot(X.T,X)
    secondpart = np.dot(X.T,Y)
    
    theta =  np.linalg.pinv(firstpart)*secondpart
    
    return theta

```

```python
theta= get_theta(X_,Y)

print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label='prediction')
plt.title("normalised data")
plt.legend()
plt.show()
```
